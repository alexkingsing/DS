```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Subplots

```python
# we can create subplots to display multiple graphs simultaneously in a single output

# creating placeholder arrays
line = np.arange(-5,5)
quadratic = line ** 2
cubic = line ** 3
fourth = line ** 4
```

```python
plt.figure(figsize=(9,5))
# defining subplot schema. This basically creates a grid for the graphs
    ## subplot needs (number of rows, number of columns, positional argument)
        ### number of rows for the graphs
        ### number of columns for the graphs
        ### position of the graph to create

# this creates the first subplot
plt.subplot(2,2,1)
plt.plot(line, '-bo')
plt.gca().yaxis.set_label_text("Value")

# this creates the second plot at position 2 ( the 2 at the end )
plt.subplot(2,2,2,)
plt.plot(quadratic, '-ro')
plt.gca().yaxis.set_label_text("Value")

# this creates the third plot at position 3 ( the 4 at the end )
plt.subplot(2,2,3)
plt.plot(cubic, '-go')
plt.gca().yaxis.set_label_text("Value")

# this creates the fourth plot at position 4 ( the 4 at the end )
plt.subplot(2,2,4)
plt.plot(fourth, '-bo')
plt.gca().yaxis.set_label_text("Value")
```

```python
# fi we want the figure to share axis, we need to store the subplots
plt.figure(figsize=(10,8))

ax1 = plt.subplot(2,2,1)
plt.plot(line, '-bo')
ax2 = plt.subplot(2,2,2, sharey=ax1)
plt.plot(quadratic, '-ro')
ax3 = plt.subplot(2,2,3, sharey=ax1)
plt.plot(cubic, '-go')
ax4 = plt.subplot(2,2,4, sharey=ax1)
plt.plot(fourth, '-bo')
plt.suptitle("THE BIG PLOT")
```

```python
# we can use the subplots function get process many graphs at once with the same axes

# this unpacks the figures and the different graphs.
    ## notice the graphs is a tuple of tuples
    ## the subplots arguments are the shape of the grid and if axis are to be shared
    # you can also directly set the figsize
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(10,8))
ax1.plot(line, '-bo')
ax2.plot(quadratic, '-ro')
ax3.plot(cubic, '-go')
ax4.plot(fourth, '-bo')

fig.text(0.5, 0.04, 'common xlabel', ha='center', va='center')
fig.text(0.06, 0.5, 'common ylabel', ha='center', va='center', rotation='vertical')
```

```python
# we can use gridspecs to create custom grids
import matplotlib.gridspec as gridspecs

plt.figure(figsize=(10,8))
gspec = gridspecs.GridSpec(3,3)

# creating a box located at first row, and uses columns 2 and 3
top_hist = plt.subplot(gspec[0,1:])
# creating a box located at second row, and uses the first column
mid_hist = plt.subplot(gspec[1:,0])
# creating a box located at second to last row, and uses columns 2 and 3
bottom_hist = plt.subplot(gspec[1:,1:])

#creating random distributions
y = np.random.normal(loc=0.0, scale=1.0, size=10000)
x = np.random.random(size=10000)

#creating a scatter of the data
bottom_hist.scatter(x, y, color='red')
#creating a normal hist
top_hist.hist(x, bins=100)
#creating a sideways hist
    ## the assignment at the end prevents the last bit of code from spamming the results of the assignment below
_=mid_hist.hist(y, bins=100, orientation='horizontal')
```

# Histogram

```python
# a histogram is a bar chart of frequency distributions

# for this example we'll create a subplot with 4 histograms
    ## the histograms will be based in the normal distribution, but with different sample sizes

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2, 2, sharex=True, figsize=(10,8))
axs = [ax1,ax2,ax3,ax4]

for n in range(len(axs)):
    # sample size = 10 to the power of N+1
    sample_size = 10**(n+1)
    # retrieving a sample using numpy,s normal generator
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    # plotting the histogram
    axs[n].hist(sample)
    # setting titles
    axs[n].set_title(f"Sample size:{sample_size}")

fig.text(0.5, 0.05, "Value")
fig.text(0.05, 0.5, "Frequency", rotation=90)

# for some rason this only works with jpg...
plt.savefig("week3.jpg")
```

```python
# doing the same but changing the amount of bins the histogram makes

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2, 2, sharex=True, figsize=(10,8))
axs = [ax1,ax2,ax3,ax4]

for n in range(len(axs)):
    # sample size = 10 to the power of N+1
    sample_size = 10**(n+1)
    # retrieving a sample using numpy,s normal generator
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    # plotting the histogram
    axs[n].hist(sample, bins=50)
    # setting titles
    axs[n].set_title(f"Sample size:{sample_size}")

fig.text(0.5, 0.05, "Value")
fig.text(0.05, 0.5, "Frequency", rotation=90)
```

# Box Plots

```python
# we will create 3 variables too represent box plots
normal_samp = np.random.normal(loc=0, scale=1, size=10000)
random_samp = np.random.random(size=10000)
gamma_samp = np.random.gamma(2, size=10000)

#uploading to a df
df = pd.DataFrame({"normal":normal_samp,
                  "random":random_samp,
                  "gamma":gamma_samp})
df.head()
```

```python
# we can use pandas describe to find summary statistics
df.describe()
```

```python

#assigning underscore prevents the output from being spammed
# creating the boxplot. whis = range means the ends are the max and min of the dist
_=plt.boxplot(df["normal"], whis='range')
```

```python
# we can pass a list of arrays to get all plots at once. We cant plot them as if they were lines

_ = plt.boxplot([df["normal"], df["gamma"], df["random"]], whis='range')
```

# Heatmaps

```python
# heatmaps are created through the use of 2 histograms
plt.figure(figsize=(10,8))
gdspecs = gridspecs.GridSpec(2,2)

top_left = plt.subplot(gdspecs[0,0])
top_right = plt.subplot(gdspecs[0,1])
bottom = plt.subplot(gdspecs[1,:])

top_left.hist(normal_samp, color="red", bins=100)
top_right.hist(random_samp, color="blue", bins=100)
_=bottom.hist2d(normal_samp, random_samp, bins=100, cmap="seismic")
```
