```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Subplots

```python
# we can create subplots to display multiple graphs simultaneously in a single output

# creating placeholder arrays
line = np.arange(10)
quadratic = line**2
```

```python
# defining subplot schema. This basically creates a grid for the graphs
    ## subplot needs (number of rows, number of columns, positional argument)
        ### number of rows for the graphs
        ### number of columns for the graphs
        ### position of the graph to create

# this creates the first subplot
plt.subplot(1,2,1)
plt.plot(line, '-bo')
plt.gca().yaxis.set_label_text("Value")

# this creates the second plot at position 2 ( the 2 at the end )
plt.subplot(1,2,2)
plt.plot(quadratic, '-ro')
plt.gca().yaxis.set_label_text("Value")
```

```python
plt.subplot(1,2,1)
plt.plot(line, '-go')
plt.axis([0,10,0,10])
```
