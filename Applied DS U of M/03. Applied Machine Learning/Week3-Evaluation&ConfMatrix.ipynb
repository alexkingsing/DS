```python
# to do this in sklearn we just need to import the appropiate measures directly

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# all of this scores take the same inputs as the confusion matrix, for example:
accuracy_score(y_test, y_predicted)

'''
In the multiclass scenario we can also calculate Macro-average and Micro-average precision/recalls. 

Macro-Average is the average of the overall metric (for example precision). To obtain this we can simply average the score of the metric for all classes.
Micro-average is the value obtained by considering ALL DATA INSTANCES as being equally important. 
For example, if we had 100 data points and we correctly predicted 90 of them, the Micro-average will be  90/100.

If the classes are similarly represented, either metric is ok.
If the classes are unevenly represented, then:
    - To give more importance to highly represented classes, use micro-average
    - To give more importance to scarcely represented classes, use macro-average
'''

# to use micro or macro averages we just specify the appropiate argument.
    # we can use F1, recall, etc
precision_score(y_test, y_predicted, average="macro")
precision_score(y_test, y_predicted, average="micro")
```
