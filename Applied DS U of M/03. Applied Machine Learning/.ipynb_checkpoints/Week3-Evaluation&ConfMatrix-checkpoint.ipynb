```python
# to do this in sklearn we just need to import the appropiate measures directly

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# all of this scores take the same inputs as the confusion matrix, for example:
accuracy_score(y_test, y_predicted)

# we can also import a general classification report function if we want all of them in a single report
from sklearn.metrics import classification_report

# this also takes the same input, but also allows some additional labeling
classification_report(y_test, y_predicted, target_names=["not 1", "1"]
```
