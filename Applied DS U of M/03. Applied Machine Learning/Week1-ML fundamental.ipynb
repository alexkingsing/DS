```python
import scipy as sp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

# Fundamentals of Machine Learning

```python
#uploading sample DF
df = pd.read_csv("data/fruit_data_with_colors.txt", sep="\t")
df.head()
```

## Creating a train-test sample

```python
# creating a dictionary with the labels and names
    # using zip to create relevant tuples
    # transforming into dictionary for easier use
names = dict(zip(df["fruit_label"].unique(), df["fruit_name"].unique()))
names
```

```python
# getting relevant columns for INDEPENDET VARIABLES
X = df[["mass", "width", "height"]]

# getting relevant column for DEPENDENT VARIABLES
y = df["fruit_name"]

# by default this uses 75-25 splits
#unpacking results from the splits
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0) 

# x-train matches y_train results
# x_test is to be used after training to see if it matches y_test
```

```python
# creating  a pairplot to visualize groupings
    # colored based on label
    
# what we are looking for here is CLEAR CLUSTERS of paired data that would signal good info for the classifier
    
pd.plotting.scatter_matrix(df[["height", "width", "mass", "color_score"]], figsize=(14,10), c=df["fruit_label"], s=80);
```

```python
# creating a 3d plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(16,12))
ax = fig.add_subplot(1,1,1, projection="3d")
ax.scatter(df["width"], df["height"], df["color_score"], c=df["fruit_label"], marker="o", s=100)
ax.set_xlabel("width")
ax.set_ylabel("height")
ax.set_zlabel("color_score");
```

# K-NEAREST NEIGHBORS

```python
# import the necessary modules to create an instance of the classifier
    # this is the module for KNN-CLASSIFIER (results a categorical Y)
from sklearn.neighbors import KNeighborsClassifier

    # this sets the number of neighborst to be used
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
# to train the method we pass both training sets, Xs & Ys
knn.fit(X_train, y_train)
```

```python
# to estimate the accuracy of the trained model, using the test sets, we can use the score method
    # this returns a decimal which represents the proportion of data that was accuracutely predicted by the model
knn.score(X_test, y_test)
```

```python
# to actually use the model to PREDICT, we use the predct method and we pass an array of the dimensions used in the training set
    # in this case the KNN algorithm requires a 2D array of data to generate a prediction
    # the 2D array can be simulated using a list of lists (similar to np arrays)
prediction = knn.predict([[20,5,10]])

# passing the resulting array into the dict of names to  label the result
prediction
```

```python
# THE MODEL'S ACCURACY, IN THIS CASE, DECREASES WITH NUMBER OF NEIGHBORS

# the model's accuracy can be tested programatically
accuracy = []
for k in range(1,21):
    # we create a KNN instance
    clf = KNeighborsClassifier(n_neighbors=k)
    # we fit the model to the data
    clf.fit(X_train, y_train)
    # we calculate the accuracy and append it to our records
    accuracy.append(clf.score(X_test, y_test))

plt.figure(figsize=(16,8))
plt.plot(accuracy, color="red", marker="o")
plt.xlabel("K")
plt.ylabel("Accuracy")
plt.xticks(list(range(1,21)));
```
